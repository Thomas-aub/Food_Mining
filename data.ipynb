```python
import pandas as pd


# Load the dataset
df1 = pd.read_csv("data/row_data/recipes.csv")
df2 = pd.read_csv("data/row_data/terms.csv")

display(df1.head())
display(df2.head())

```

```python
merged_df = pd.merge(df1, df2, on='id', how='inner')

with pd.option_context('display.max_colwidth', None):
    display(merged_df.head(1))
```

```python
df = merged_df[["name_x", "nutrition", "n_steps", "ingredients_y", "minutes"]]
df.to_csv("data/clean_data/recipes_cleaned.csv", index=False)
df.head()
```
